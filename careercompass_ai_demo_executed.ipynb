## 1. Problem Statement

Students frequently struggle with:

- choosing realistic career paths,
- understanding skill gaps,
- preparing for internships,
- improving resumes and interview readiness,
- converting vague goals into a time-bound action plan.

Generic chatbots often provide paragraph advice. CareerCompass AI is designed to produce a **structured, explainable, multi-agent career report**.

## What Makes This Different?

| Generic Chatbot | CareerCompass AI |
|---|---|
| One broad recommendation | Compares multiple career paths side by side |
| Paragraph advice | Structured report with sections |
| Limited transparency | Explainable Suitability Index |
| Generic suggestions | Deadline-aware internship sprint roadmap |
| Passive response | Human-in-the-loop approval before final roadmap |

### Core Differentiator

> **Career Path Comparison Matrix + Deadline-Aware Internship Sprint Plan**

## 2. Multi-Agent Architecture

### Visual Workflow

```text
Student Profile
      ↓
Privacy & Safety Agent
      ↓
Coordinator Agent
      ↓
Career Path Comparator
      ↓
Skill Gap Analysis
      ↓
Roadmap Planner
      ↓
Human Approval
      ↓
Career Report
```

### Architecture Diagram

![CareerCompass AI Architecture](assets/architecture_diagram.svg)

## Agent Responsibilities

| Agent | Responsibility |
|---|---|
| Privacy & Safety Agent | Masks sensitive information and validates user input |
| Coordinator Agent | Orchestrates the workflow and composes the final report |
| Career Comparator Agent | Scores multiple career paths using deterministic criteria |
| Skill Gap Agent | Finds missing skills and readiness gaps |
| Roadmap Planner Agent | Creates a deadline-aware internship preparation plan |
| Human Approval Agent | Lets the user review or adjust the recommendation before final output |

## 3. Course Concepts Demonstrated

## 9. Limitations

- This MVP expects a structured profile dictionary, not fully free-form natural language input.
- It does not scrape live internship listings.
- It does not guarantee job or internship outcomes.
- It recommends among configured career paths only.
- Recommendations should be reviewed by a human mentor, faculty member, or career counselor.
